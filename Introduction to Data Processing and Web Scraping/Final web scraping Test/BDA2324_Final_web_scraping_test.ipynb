# Goal-1

```python
def get_discription(page):
  #print(page.find('div',itemprop=re.compile('description')).p.text)
  #discription = page.find('div',itemprop=re.compile('description')).span.text.replace('\xa0','')
  dicr =page.find('div',itemprop=re.compile('description')).p
  if dicr == None:
    discription = page.find('div',itemprop=re.compile('description')).span.text.replace('\xa0','')
  else:
    discription = page.find('div',itemprop=re.compile('description')).p.text
  return discription
```

```python
def get_price(page):
  price = page.find('span',id=re.compile('productPrice')).text
  count=[]
  for i in price.split(' '):

    if i!='':
      if i!='\n':
        count.append(i)

  amount = float(count[0].replace('\n','').replace('$',''))
  return amount
```

```python
url = 'https://americanmadestore.us/products/amish-wedding-apple-butter-32oz?_pos=1&_sid=6ebdb3865&_ss=r'
p = requests.get(url)
page = bs4.BeautifulSoup(p.content,'lxml')
title = page.h1.text
discription = get_discription(page)[:100]
price = get_price(page)
weight = page.find('div',class_=re.compile('selector-wrapper'))
if weight == None:
  weight = False
else:
  weight = True

print((title, price, discription,weight))
```

```python
def scrape_product_details(url):
  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content,'lxml')
  title = page.h1.text
  discription = get_discription(page)[:100]
  price = get_price(page)
  weight = page.find('div',class_=re.compile('selector-wrapper'))
  if weight == None:
    weight = False
  else:
    weight = True

  return (title, price, discription,weight)
```

# Goal-2

```python
def get_discription(page):
  #print(page.find('div',itemprop=re.compile('description')).p.text)
  #discription = page.find('div',itemprop=re.compile('description')).span.text.replace('\xa0','')
  dicr =page.find('div',itemprop=re.compile('description')).p
  if dicr == None:
    discription = page.find('div',itemprop=re.compile('description')).span.text.replace('\xa0','')
  else:
    discription = page.find('div',itemprop=re.compile('description')).p.text
  return discription
```

```python
def get_price(page):
  price = page.find('span',id=re.compile('productPrice')).text
  count=[]
  for i in price.split(' '):

    if i!='':
      if i!='\n':
        count.append(i)

  amount = float(count[0].replace('\n','').replace('$',''))
  return amount
```

```python
def scrape_product_details(url):
  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content,'lxml')
  title = page.h1.text
  discription = get_discription(page)[:100]
  price = get_price(page)
  weight = page.find('div',class_=re.compile('selector-wrapper'))
  if weight == None:
    weight = False
  else:
    weight = True

  return (title, price, discription,weight)
```

```python
def search_all_details(url,simulate=False):
  total_itemsper_page = 12
  fetched_ads_count = 0

  base_url = 'https://americanmadestore.us/'
  print(f"Main Url: {url}")
  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content)
  all_advertisements = page.find_all('div',class_='desc')

  result = []

  for ad in all_advertisements:
    #print('First')
    #print(ad)
    if fetched_ads_count<total_itemsper_page:
      ad_url = base_url+ad.h5.a['href']

      if not simulate:
        time.sleep(1)
        print(f'Product Number {fetched_ads_count+1}. ',end='')
        print(ad_url)
        result.append(scrape_product_details(ad_url))

        fetched_ads_count += 1


      else:
          print(f'Fetching {ad_url}')

  temp_df = pd.DataFrame(result,  columns=['Title','Price','Description','Weightable'])
  #temp_df.set_index('Stock number', inplace=True)
  #temp_df.to_excel(f'{item_name}_ads_{date_of_scraping}.xlsx', index=False)
  #print('Excel File Saved Successfully!')
  return temp_df
```

```python
search_URL = 'https://americanmadestore.us/search?type=product&q=apple'
all_products = search_all_details(search_URL)
```

```python
def dataframe_lower_price(df, price_value):
  lower_price = df[df['Price']<price_value]
  return lower_price
```
