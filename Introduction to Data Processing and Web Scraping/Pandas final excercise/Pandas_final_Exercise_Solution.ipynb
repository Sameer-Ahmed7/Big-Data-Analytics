# Excercise # 01 (Transaction Data Analysis with Pandas)

```python
import pandas as pd
```

```python
t = pd.read_csv('Transactions.csv', dtype={'committer':'str','receiver':'str'}, header=None,names = ['id','amount','committer','receiver','type'])
```

```python
t.head(5)
```

## Bank Transfer Transactions:

```python
def bank_transfer_transactions():
  bank_transfer_t = t[t['type']=='bank transfer']
  return bank_transfer_t
```

```python
bank_transfer_t = bank_transfer_transactions()
```

```python
bank_transfer_t
```

## User-Specific Transactions:

```python
def user_spec_trans(committer='323'):
  user_spec_t = t[t['committer']==committer]
  return user_spec_t
```

```python
user_spec_t = user_spec_trans()
```

```python
user_spec_t
```

## Multiple Specific Transactions:

```python
def mult_spec_trans(committer_1='323',committer_2='194'):
  mult_spec_t = t[(t['committer']==committer_1) | (t['committer']==committer_2)]
  return mult_spec_t
```

```python
mult_spec_t = mult_spec_trans()
```

```python
mult_spec_t
```

## Aggregated Transaction Amounts:

```python
def agg_trans_amount():
  value = t[(t['type']=='credit card') | (t['type']=='cheque')]
  agg_t_amount = value['amount'].sum()
  return agg_t_amount
```

```python
agg_t_amount = agg_trans_amount()
```

```python
print(f'Aggregated Transaction Amounts: {agg_t_amount}')
```

## Filtered Transaction Range:

```python
def filt_trans_range():
  filt_t_range  = t[(t['type']=='bank transfer') & (t['id'].between(1,10))]
  return filt_t_range
```

```python
filt_t_range = filt_trans_range()
```

```python
filt_t_range
```

## High-Value Bank Transfers:

```python
def high_v_bank_trans():
  high_value = t[(t['type']=='bank transfer') & (t['amount']>10000)]
  return high_value
```

```python
high_value = high_v_bank_trans()
```

```python
high_value
```

# Excercise # 02 (Data Integration and Analysis with Pandas)

```python
import pandas as pd
```

```python
s.drop(columns='OLDID', inplace=True)
```

```python
replacements = {'Female':'F','Male':'M'}
s['GENDER'].replace(replacements, inplace=True)
s
```

```python
s['BIRTHDATE'] = pd.to_datetime(s['BIRTHDAY'].astype(str)+'/'+s['BIRTHMONTH'].astype(str)+'/'+s['BIRTHYEAR'].astype(str))
s.drop(columns=['BIRTHYEAR','BIRTHDAY','BIRTHMONTH'], inplace=True)
```

```python
s
```

## Committer Variables in Subjects Dataframe

```python
def committer_subject_dataframe():
  s['committer'] = False
  for comm_uniq in list(t_2['committer'].unique()):
    if s['ID'][s['ID']==comm_uniq].count() >=1:
      s['committer'][s['ID']==comm_uniq] = True
    else:
      s['committer'][s['ID']==comm_uniq] = False
```

```python
committer_subject_dataframe()
```

```python
s.head()
```

## Tax Variables in Transection Dataframes

```python
t_2.head(5)
```

```python
import numpy as np
```

```python
def tax_transection_dataframe():
  t_2['tax']=0.0
  t_2['tax'] = np.where(t_2['type'] == 'bank transfer', t_2['amount'] * 0.02,
                    np.where(t_2['type'] == 'credit card', t_2['amount'] * 0.03, 0.0))
```

```python
tax_transection_dataframe()
```

```python
t_2.head(5)
```

## Transection Counts and Tax Amounts

```python
def trans_count_tax_amount():
  t_count_tax_amount = t_2.groupby('receiver').agg({'id': 'count', 'tax': 'sum'}).reset_index()
  t_count_tax_amount.rename(columns={'receiver':'receiver_id','id': 'total_received_transactions', 'tax': 'total_taxes_paid'}, inplace=True)
  return t_count_tax_amount
```

```python
t_count_tax_amount = trans_count_tax_amount()
```

```python
t_count_tax_amount
```

## Export Tax Summary


```python
def save_tax_summary(t_count_tax_amount,file_name='tax_summary.xlsx'):
  total_tax = []
  for id in list(s['ID'].values):
    if list(t_count_tax_amount['total_taxes_paid'][t_count_tax_amount['receiver_id']==id].values) == []:
      total_tax.append(0.0)
    else:
      total_tax.append(list(t_count_tax_amount['total_taxes_paid'][t_count_tax_amount['receiver_id']==id].values)[0])


  data = {
      "ID number":s['ID'].values,
      "first name":s['FIRSTNAME'].values,
      "last name":s['LASTNAME'].values,
      "total amount of taxes paid":total_tax,

  }
  df = pd.DataFrame(data)

  # Save the DataFrame to an Excel file
  output_file = file_name
  df.to_excel(output_file, index=False)

  print(f"DataFrame saved to {output_file}")
```

```python
save_tax_summary(t_count_tax_amount)
```
