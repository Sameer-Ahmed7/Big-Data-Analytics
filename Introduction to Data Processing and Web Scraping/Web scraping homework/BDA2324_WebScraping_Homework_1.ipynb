## Goal-1

```python
import requests
import bs4
import re
from datetime import datetime

import pytz

import time
import pandas as pd

```

### Get Current Date Function

```python
def get_date():
  tz = pytz.timezone('Europe/Rome')
  current_dateTime = datetime.now()
  date_of_scraping = f'{current_dateTime.day}-{current_dateTime.month}-{current_dateTime.year}'
  return date_of_scraping
```

### Scrap Product Function

```python
def scrape_product(url):
  print(f'Scraping data from: {url}')

  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content,'lxml')
  title = page.h1.text[1:]
  seller = page.find('div',class_=re.compile('ux-seller-section__item--seller')).find('span',class_=re.compile('ux-textspans'))
  if seller == None:
    seller_guy = ""
  else:
    seller_guy = seller.text

  currency = page.find('div',class_=re.compile('x-price-primary')).find('span',class_=re.compile('ux-textspans')).text.split(' ')[0]

  if currency.upper() != "EUR":
    selling_price = float(page.find('div',class_=re.compile('x-price-approx')).find('span',class_=re.compile('x-price-approx__price')).text.split(' ')[1].replace(',','.'))
  else:
    selling_price = float(page.find('div',class_=re.compile('x-price-primary')).find('span',class_=re.compile('ux-textspans')).text.split(' ')[1].replace(',','.'))

  date_of_scraping = get_date()

  return (title, seller_guy, selling_price, date_of_scraping)

```

### Search By Keyword Function

```python
def search_by_keyword(item_name, max_results_number = 50, simulate=False):
  fetched_ads_count = 0
  item_name = item_name.replace(" ", "")
  url = f'https://www.ebay.it/sch/i.html?_from=R40&_trksid=p4432023.m570.l1313&_nkw={item_name}&_sacat=0'
  print(f"Main Url: {url}")
  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content)
  ads_number_string = page.find('h1',class_=re.compile('srp-controls__count-heading')).find('span',class_=re.compile('BOLD')).text
  ads_number_string = ads_number_string.replace('.','')


  ads_number = int(ads_number_string)
  print(f'Found {ads_number} results. Scraping started.')

  page_count = ads_number//60+1





  for current_page in range(1, page_count+1):

    print(f'Fetching results from page {current_page}')

    url = f'https://www.ebay.it/sch/i.html?_from=R40&_trksid=p4432023.m570.l1313&_nkw={item_name}&_sacat=0&_pgn={current_page}'
    p = requests.get(url)
    page = bs4.BeautifulSoup(p.content)
    all_advertisements = page.find_all('a',class_='s-item__link')

    result = []

    #print(all_advertisements)
    count=0
    for ad in all_advertisements:
      #print(ad['href'])
      if count==0:
        pass
      else:
        ad_url = ad['href']
        if not simulate:
          time.sleep(1)
          print(f'Product Number {fetched_ads_count+1}. ',end='')
          result.append(scrape_product(ad_url))

          fetched_ads_count += 1
          if fetched_ads_count == max_results_number:
            print('Reached the maximum number of ads to scrape. Exiting')
            temp_df = pd.DataFrame(result,  columns=['Name_of_the_product', 'seller', 'selling_price', 'date_of_scraping'])
            #temp_df.set_index('id', inplace=True)
            date_of_scraping = get_date()

            temp_df.to_excel(f'{item_name}_ads_{date_of_scraping}.xlsx', index=False)
            print('Excel File Saved Successfully!')
            return temp_df

        else:
            print(f'Fetching {ad_url}')

      count+=1
    temp_df = pd.DataFrame(result,  columns=['Name_of_the_product', 'seller', 'selling_price', 'date_of_scraping'])

    date_of_scraping = get_date()

    temp_df.to_excel(f'{item_name}_ads_{date_of_scraping}.xlsx', index=False)
    print('Excel File Saved Successfully!')
    return temp_df
```

```python
def dataframe_specific_string(df,specific_string):
  df_specific_string = df[df['Name_of_the_product'].str.contains(specific_string)]
  return df_specific_string
```
