# Scraping data from National Hardware Website


## Import Libraries

```python
import requests
import bs4
import re

import time
import pandas as pd
```

## Goal 1

### Get Discription Function

```python
def get_discription(page):
  filtered_data = ''
  #print(page.find('ul',class_=re.compile('list-bulleted')))
  non_filtered_disc = page.find('ul',class_=re.compile('list-bulleted')).findAll('li')
  for disc in non_filtered_disc:
    non_filtered = disc.text
    filtered_t = non_filtered.replace('\t','')
    filtered_n = filtered_t.replace('\n','')
    filtered_r = filtered_n.replace('\r','')
    filtered_data+=filtered_r
    filtered_data+='. '
  return filtered_data
```

### Scrap Product Function

```python
def scrape_product_details(url):
  print(f'Scraping data from: {url}')

  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content,'lxml')
  title = page.h1.text
  discription = get_discription(page)[:100]
  stock_number = page.find('div',class_=re.compile('kicker')).text.replace('Stock # ','')

  return (stock_number,title, discription)
```

#### Execute Function

```python
url = 'https://www.national-hardware.com/p/120-chair-braces?model=N234-617'
ad = scrape_product_details(url)
print(ad)
```

## Goal-2

### Get Discription Function

```python
def get_discription(page):
  filtered_data = ''
  #print(page.find('ul',class_=re.compile('list-bulleted')))
  non_filtered_disc = page.find('ul',class_=re.compile('list-bulleted')).findAll('li')
  for disc in non_filtered_disc:
    non_filtered = disc.text
    filtered_t = non_filtered.replace('\t','')
    filtered_n = filtered_t.replace('\n','')
    filtered_r = filtered_n.replace('\r','')
    filtered_data+=filtered_r
    filtered_data+='. '
  return filtered_data
```

### Scrap Product Function

```python
def scrape_product_details(url):
  print(f'Scraping data from: {url}')

  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content,'lxml')
  title = page.h1.text
  discription = get_discription(page)[:100]
  stock_number = page.find('div',class_=re.compile('kicker')).text.replace('Stock # ','')

  return (stock_number,title, discription)
```

### Search By All Details Function

```python
def search_all_details(url,simulate=False):

  fetched_ads_count = 0

  base_url = 'https://www.national-hardware.com'
  print(f"Main Url: {url}")
  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content)
  all_advertisements = page.find_all('div',class_='search-results-product__heading')

  result = []

  for ad in all_advertisements:
    ad_url = base_url+ad.find('a')['href']
    if not simulate:
      time.sleep(1)
      print(f'Product Number {fetched_ads_count+1}. ',end='')
      result.append(scrape_product_details(ad_url))

      fetched_ads_count += 1


    else:
        print(f'Fetching {ad_url}')

  temp_df = pd.DataFrame(result,  columns=['Stock number','Title','Description'])
  temp_df.set_index('Stock number', inplace=True)
  #temp_df.to_excel(f'{item_name}_ads_{date_of_scraping}.xlsx', index=False)
  #print('Excel File Saved Successfully!')
  return temp_df
```

#### Execute Function

```python
search_URL = 'https://www.national-hardware.com/search?FilterCriteria.Query=chair'
all_products = search_all_details(search_URL)
```

```python
def search_by_keyword(item_name,simulate=False):

  fetched_ads_count = 0

  base_url = 'https://www.national-hardware.com'

  item_name = item_name.replace(" ", "")
  url = f'https://www.national-hardware.com/search?FilterCriteria.Query={item_name}'
  print(f"Main Url: {url}")
  p = requests.get(url)
  page = bs4.BeautifulSoup(p.content)
  all_advertisements = page.find_all('div',class_='search-results-product__heading')

  result = []

  for ad in all_advertisements:
    ad_url = base_url+ad.find('a')['href']
    if not simulate:
      time.sleep(1)
      print(f'Product Number {fetched_ads_count+1}. ',end='')
      result.append(scrape_product_details(ad_url))

      fetched_ads_count += 1


    else:
        print(f'Fetching {ad_url}')

  temp_df = pd.DataFrame(result,  columns=['Stock number','Title','Description'])
  temp_df.set_index('Stock number', inplace=True)
  #temp_df.to_excel(f'{item_name}_ads_{date_of_scraping}.xlsx', index=False)
  #print('Excel File Saved Successfully!')
  return temp_df
```

```python
def dataframe_specific_string(df,specific_string):
  specific_string = specific_string.title()
  df_specific_string = df[df['Title'].str.contains(specific_string)]
  return df_specific_string
```
